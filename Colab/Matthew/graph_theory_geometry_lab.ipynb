```python
import networkx as nx
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Introduction / Basics
Running through the basics of the Networkx library / getting accustomed to it

```python
#Undirected graph
G = nx.Graph()
```

```python
#Add edge between point 1 and point 2
#This will also create the nodes specified if they are not already defined
#For a directed graph, this will go from node A to node B
G.add_edge(1,2)

#Can also just add an individual node
G.add_node(5)
```

```python
#Can also add / specify weight to the edge
G.add_edge(2,3, weight = 0.9)
```

```python
nx.draw_spring(G, with_labels = True)
plt.show()
```

```python
#Can also define a graph using edge lists
edge_list = [(1,2), (3,4), (2,3),(5,6),(6,2), (5,3), (7,1), (4,7)]
G=nx.from_edgelist(edge_list)
nx.draw_spring(G, with_labels = True)
plt.show()
```

```python
#Can define a graph using an adjacency matrix
#Can create this matrix with numpy

aj_matrix = np.array([[0,1,1],
                      [1,0,0],
                      [1,0,0]])
G = nx.from_numpy_array(aj_matrix)
nx.draw_spring(G, with_labels = True)
plt.show()
```

```python
#Different ways of drawing graphs
nx.draw_spring()
nx.draw_circular()
nx.draw_shell()
nx.draw_spectral()
nx.draw_random()
nx.draw_planar()
```

```python
#Random graphs
G = nx.erdos_renyi_graph(50,0.15)
nx.draw_spring(G, with_labels = True)
plt.show()
```
